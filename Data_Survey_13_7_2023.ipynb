##plot แสดงเพศและอายุเฉลี่ยของผู้ตอบแบบสอบถาม

```python
data3.groupby('เพศ').mean()
```

```python
data31 = data3.groupby('เพศ').mean().reset_index()[['เพศ','อายุ']]
data31.sort_values(by=['อายุ'])
```

```python
sns.barplot(x='เพศ', y='อายุ', data=data31)
```

##plot แสดงเพศและอุณหภูมิเฉลี่ยของผู้ตอบแบบสอบถาม

```python
data3.groupby('เพศ').mean()
```

```python
data32 = data3.groupby('เพศ').mean().reset_index()[['เพศ','Temp']]
data32.sort_values(by=['Temp'])
```

```python
sns.barplot(x='เพศ', y='Temp', data=data32)
```

##plot แสดงพื้นที่ทำแบบสอบถามและอุณหภูมิเฉลี่ยของผู้ตอบแบบสอบถาม

```python
data3.groupby('พื้นที่ทำแบบสอบถาม').mean()
```

```python
data33 = data3.groupby('พื้นที่ทำแบบสอบถาม').mean().reset_index()[['พื้นที่ทำแบบสอบถาม','Temp']]
data33.sort_values(by=['Temp'])
```

```python
sns.barplot(x='Temp', y='พื้นที่ทำแบบสอบถาม', data=data33)
```
