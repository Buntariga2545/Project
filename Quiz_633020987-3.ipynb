##สร้าง plot แสดงค่าเฉลี่ยของอายุ (age) ของตัวอย่างในแต่ละเพศ (sex)

```python
adult.groupby('sex').mean()
```

```python
adult11 = adult.groupby('sex').mean().reset_index()[['sex','age']]
adult11.sort_values(by=['age'])
```

```python
sns.barplot(x='sex', y='age', data=adult11)
```

##สร้าง plot แสดงค่าเฉลี่ยของจำนวนชั่วโมงทำงานต่อสัปดาห์ (hours-per-week) ของตัวอย่างในแต่ละเชื้อชาติ (race)

```python
adult.groupby('race').mean()
```

```python
adult12 = adult.groupby('race').mean().reset_index()[['race','hours-per-week']]
adult12.sort_values(by=['hours-per-week'])
```

```python
sns.barplot(x='hours-per-week', y='race', data=adult12, palette="viridis")
```
